Lesson 01: Differential Calculus
In this lesson, you will discover what is differential calculus or differentiation.

Differentiation is the operation of transforming one mathematical function to another, called the derivative. The derivative tells the slope, or the rate of change, of the original function.

For example, if we have a function f(x)=x**2
, its derivative is a function that tells us the rate of change of this function at . The rate of change is defined as:
 
f'(x)=(f(x+dx)-f(x))/dx

for a small quantity .

Usually we will define the above in the form of a limit, i.e.,

 
 

to mean  should be as close to zero as possible.

There are several rules of differentiation to help us find the derivative easier. One rule that fits the above example is 
 d/dx X^n= n X^(n-1)


. Hence for 
, we have the derivative 
.

We can confirm this is the case by plotting the function 
 computed according to the rate of change together with that computed according to the rule of differentiation. The following uses NumPy and matplotlib in Python:

```python
#calculus 1 differential

def f(x):
    return (x**2)
```

```python
plt.style.use("ggplot")
x=np.linspace(-10,10,500)
y=f(x)

#ploting fx


fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(2,2,1) #subplot positiion (nrow, ncol, n_indext (start from 1 upper left corner, increase to right places))
ax.plot(x,y)
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.set_title("y=f(x)")

#plotiing f'(x) using rate
dx=0.0001
y1=(f(x+dx)-f(x))/dx

#using role
y2=2*x

#plotting y1 and y2

ax=fig.add_subplot(2,2,4)
ax.plot(x,y1, c='r', label="rate", alpha=0.5)
ax.plot(x,y2, c='b', label="role", alpha=0.5)
ax.set_xlabel("x")
ax.set_ylabel("f '(x)")
ax.legend()
ax.set_title("rate and rule")



plt.show()
```

lesson 02 INtegration
====================

intergration is reverse of derifative term. if derivative tells us about slope or the rate of change,
then the integration tells us about the "SUM" of result in every point of derivative, a.k.a area under curve of function,
since when the curve is a function then the integrative is area uder that curve.

for instance:

f(x)=x**2
f'(X)=2x

if from x=-10 to 10 with interval=0.1 -->what the result 

integration of AUC f(x)= x**2,
then, fintegral= f(-10)*0.1 + f(-9.9)*0.1.....+f(10)*0.1 #sum fx + interval


```python
def f(x):
    return(2*x)


x=np.arange(-10,10, 0.1) # np.arange(min, max, interval)
dx=0.1

fx=f(x)*dx

#running sum
y=fx.cumsum()

#plot
plt.plot(x,y)
plt.show()
```

```python
#exercise
#f(x)=3x**2-4*x

def f(x):
    return(3*(x**2)-(4*x))

x=np.arange(-10,10,0.1)
dx=0.1

fx=f(x)*dx
y=f(x) #native
y1=fx.cumsum() #antiderivative or integrale based on sum theory

#rule 
def z(x):
    return((x**3))-(2*(x**2))

y2=z(x)



#plotting native function (derivative)
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(2,2,1)
ax.plot(x,y, label="derivative", c='b')
ax.legend()

#plotting integrative for native function based on sum theory
ax=fig.add_subplot(2,2,2)
ax.plot(x,y1, label="antiderivative or integral based on area under curve", c='r')
ax.plot(x,y2, label="intergral based on rule can't show actual value due tu 'c value' wasn't define", c='green')
ax.legend()

plt.show()
```
